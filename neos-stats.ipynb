```python
import urllib
import re

stat_kinds = {
  'name': 'Solver Names',
  'input': 'Solver Inputs',
  'catagory': 'Solver Catagories',
  'interface': 'Interfaces',
  'domain': 'Domains'
}

def get_stats(start_date, end_date, stat_kind):
  data = {
    'start': start_date, 'end': end_date,
    stat_kind: stat_kinds[stat_kind]
  }
  f = urllib.urlopen("http://www.neos-server.org/neos/cgi-bin/report.cgi",
                     urllib.urlencode(data))
  html = f.read()
  f.close()
  names = []
  values = []
  sum = 0
  for m in re.finditer(r'<tr><td>([^<]*)</td><td.*>(.*)</td></tr>', html):
    value = int(m.group(2))
    names.append(m.group(1))
    values.append(value)
    sum += value
  total = int(re.search(r'([0-9]+) Total Jobs Submitted', html).group(1))
  if total != sum:
    raise Exception("total doesn't match: {} != {}".format(total, sum))
  return names, values, total

def plot_stats(start_date, end_date, stat_kind):
  names, values, total = get_stats(start_date, end_date, stat_kind)
  N = min(len(values), 10)
  ind = np.arange(N)
  fig = plt.figure()
  ax = fig.add_subplot(111)
  fig.autofmt_xdate()
  ax.bar(ind, np.array(values[0:N]))
  ax.set_xticks(ind + 0.5)
  ax.set_xticklabels(names[0:N])
    
def get_gchart_stats(start_date, end_date, stat_kind):
  names, values, total = get_stats(start_date, end_date, stat_kind)
  for i in range(min(10, len(values))):
    print "  ['{0}', {{v: {1}, f: \"{1} ({2:.2%})\"}}],".format(
      names[i], values[i], values[i] / float(total))
```
